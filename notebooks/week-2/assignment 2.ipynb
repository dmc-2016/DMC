```python
import random
gameStake = 50
cards = range(10)
class Player:
    ID=0
    Pot=0
    def __init__(self, inputID, startingPot):
        self.ID = inputID
        self.Pot=startingPot
    def play(self, dealerCard):
        playerCard = random.choice(cards)
        if playerCard >= dealerCard:
            self.Pot += gameStake
            print 'player', self.ID, 'Win', playerCard, 'vs', dealerCard
        else:
            self.Pot -= gameStake
            print 'player', self.ID, 'Lose', playerCard, 'vs', dealerCard
    def returnPot(self):
        return self.Pot
    def returnID(self):
        return self.ID
def playHand(players):
    for player in players:
        dealerCard = random.choice(cards)
        player.play(dealerCard)
def checkBalances(players):
    for player in players:
        print 'player', player.returnID(), 'has $', player.returnPot(), 'left'
        
players = []
for i in range(5):
    players.append(Player(i,500))
for i in range(10):
    print'startgame' + str(i)
    playHand(players)
    print 'game results:'
    checkBalances(players)

```
